```python
import matplotlib.pyplot as plt
import seaborn as sns

features = ['Danceability', 'Energy',
       'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness',
       'Liveness', 'Valence', 'Tempo']
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

for i, feature in enumerate(features):
    row, col = i // 3, i % 3

    sns.histplot(df[feature], bins=30, kde=True, ax=axes[row, col], color="blue")
    axes[row, col].set_title(f"Histogram of {feature}")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

for i, feature in enumerate(features):
    row, col = i // 3, i % 3

    sns.boxplot(x=df[feature], ax=axes[row, col], color="red")
    axes[row, col].set_title(f"Box Plot of {feature}")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

for i, feature in enumerate(features):
    row, col = i // 3, i % 3

    # Violin plot
    sns.violinplot(x=df[feature], ax=axes[row, col], color="purple")
    axes[row, col].set_title(f"Violin Plot of {feature}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[features].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```
